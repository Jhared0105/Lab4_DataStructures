```python
# ============================================================
# ADDITIONAL EXERCISE 1: Unstructured Data Aggregator
# Student: Dungca | ID: TUPM-25-1053
# ============================================================
 
from collections import defaultdict, Counter
 
# --- Required Student Variables ---
SEED_NUM        = 3                  # Last digit of TUPM-25-1053
STUDENT_MAJOR   = "BSME"            # Current degree program
FAVORITE_ARTIST = "FRANK ELY"       # Favorite local artist
LOCAL_HAZARD    = "EARTHQUAKE"       # Local emergency event
CONTROL_NUM     = max(3, SEED_NUM)   # max(3, 3) = 3
 
print("=== STUDENT VARIABLES ===")
print(f"SEED_NUM       : {SEED_NUM}")
print(f"STUDENT_MAJOR  : {STUDENT_MAJOR}")
print(f"FAVORITE_ARTIST: {FAVORITE_ARTIST}")
print(f"LOCAL_HAZARD   : {LOCAL_HAZARD}")
print(f"CONTROL_NUM    : {CONTROL_NUM}")
print()
 
# --- Base Application Stream ---
stream = [
    ("TUP Manila",   "BSECE"),
    ("TUP Taguig",   f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila",   f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila",   "BSECE"),
    ("TUP Visayas",  STUDENT_MAJOR),
    ("TUP Taguig",   "BSECE"),
    ("TUP Manila",   "BSECE"),
]
 
# --- Data Expansion: Add exactly (CONTROL_NUM + 3) = 6 custom tuples ---
custom_applications = [
    ("TUP Manila",   "BSECE"),          # Custom 1
    ("TUP Manila",   STUDENT_MAJOR),    # Custom 2 - BSME
    ("TUP Taguig",   STUDENT_MAJOR),    # Custom 3 - BSME
    ("TUP Visayas",  "BSECE"),          # Custom 4
    ("TUP Manila",   "BSECE"),          # Custom 5
    ("TUP Taguig",   "BSECE"),          # Custom 6
]
 
stream.extend(custom_applications)
 
print(f"=== EXPANDED APPLICATION STREAM (Total: {len(stream)}) ===")
for entry in stream:
    print(f"  {entry}")
print()
 
# --- Algorithm Phase 1: Aggregation via defaultdict ---
campus_programs = defaultdict(list)
 
for campus, program in stream:
    campus_programs[campus].append(program)
 
print("=== PHASE 1: AGGREGATED APPLICATIONS BY CAMPUS ===")
for campus, programs in campus_programs.items():
    print(f"  {campus}: {programs}")
print()
 
# --- Algorithm Phase 2: Frequency Analysis via Counter ---
print("=== PHASE 2: FREQUENCY ANALYSIS PER CAMPUS ===")
campus_counters = {}
 
for campus, programs in campus_programs.items():
    counter = Counter(programs)
    campus_counters[campus] = counter
    print(f"\n  [{campus}]")
    for program, count in counter.most_common():
        print(f"    {program}: {count} application(s)")
 
# --- Top Program at TUP Manila ---
manila_top = campus_counters["TUP Manila"].most_common(1)
print()
print("=== MOST DEMANDED PROGRAM: TUP MANILA ===")
print(f"  Program : {manila_top[0][0]}")
print(f"  Count   : {manila_top[0][1]} application(s)")
print()
 
# --- Summary Table Output ---
print("=== ASSESSMENT TABLE OUTPUT ===")
print(f"  CONTROL_NUM                        : {CONTROL_NUM}")
print(f"  Custom tuples added (CONTROL_NUM+3): {CONTROL_NUM + 3}")
print(f"  Total stream size after expand     : {len(stream)}")
print(f"  TUP Manila - Total applications   : {len(campus_programs['TUP Manila'])}")
print(f"  TUP Taguig - Total applications   : {len(campus_programs['TUP Taguig'])}")
print(f"  TUP Visayas - Total applications  : {len(campus_programs['TUP Visayas'])}")
print(f"  Manila most_common(1) result       : {manila_top}")
```
