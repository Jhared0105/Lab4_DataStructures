```python
# ADDITIONAL EXERCISE 2: Jaccard Similarity Engine
# Student: Dungca | ID: TUPM-25-1053
# ============================================================

# --- Required Student Variables ---
SEED_NUM        = 3
FAVORITE_ARTIST = "Frank Ely"
CONTROL_NUM     = max(3, SEED_NUM)  # = 3

print("=== STUDENT VARIABLES ===")
print(f"FAVORITE_ARTIST : {FAVORITE_ARTIST}")
print(f"CONTROL_NUM     : {CONTROL_NUM}")
print()

# --- Initialize Festival Lineup Set ---
festival = {
    "Ben&Ben",
    "SB19",
    "Bini",
    "Eraserheads",
    FAVORITE_ARTIST,
    "Zild",
    f"Indie Artist {CONTROL_NUM}"
}

# --- Initialize User Profile Sets ---
user_a = {"Ben&Ben", "Bini", "Maki", "Dionela", FAVORITE_ARTIST}
user_b = {"SB19", "Eraserheads", "Zild", f"Indie Artist {CONTROL_NUM}", "Parokya ni Edgar"}

# --- Data Expansion: Add exactly CONTROL_NUM (3) new artists to festival ---
new_artists = {"December Avenue", "Cup of Joe", "Nobita"}
festival.update(new_artists)

print("=== EXPANDED FESTIVAL LINEUP ===")
print(f"  {festival}")
print(f"  Total Artists in Festival: {len(festival)}")
print()

# --- Algorithm: Jaccard Similarity for User A ---
intersection_a = user_a.intersection(festival)
union_a        = user_a.union(festival)
jaccard_a      = (len(intersection_a) / len(union_a)) * 100

print("=== USER A: JACCARD SIMILARITY ===")
print(f"  User A Profile      : {user_a}")
print(f"  Intersection (A∩F)  : {intersection_a}  | Size: {len(intersection_a)}")
print(f"  Union (A∪F)         : Size = {len(union_a)}")
print(f"  Jaccard Similarity  : {len(intersection_a)} / {len(union_a)} x 100 = {jaccard_a:.2f}%")
print()

# --- Algorithm: Jaccard Similarity for User B ---
intersection_b = user_b.intersection(festival)
union_b        = user_b.union(festival)
jaccard_b      = (len(intersection_b) / len(union_b)) * 100

print("=== USER B: JACCARD SIMILARITY ===")
print(f"  User B Profile      : {user_b}")
print(f"  Intersection (B∩F)  : {intersection_b}  | Size: {len(intersection_b)}")
print(f"  Union (B∪F)         : Size = {len(union_b)}")
print(f"  Jaccard Similarity  : {len(intersection_b)} / {len(union_b)} x 100 = {jaccard_b:.2f}%")
print()

# --- Set Difference: User A's Dealbreakers ---
dealbreakers_a = user_a.difference(festival)

print("=== USER A: DEALBREAKERS ===")
print(f"  Artists User A likes but NOT in festival: {dealbreakers_a}")
print()

# --- Assessment Table Summary ---
print("=== ASSESSMENT TABLE OUTPUT ===")
print(f"  CONTROL_NUM Used                          : {CONTROL_NUM}")
print(f"  New artists added via .update()           : {new_artists}")
print(f"  Expanded Festival Size                    : {len(festival)}")
print(f"  User A - Intersection Size                : {len(intersection_a)}")
print(f"  User A - Union Size                       : {len(union_a)}")
print(f"  User A - Jaccard Similarity               : {jaccard_a:.2f}%")
print(f"  User B - Intersection Size                : {len(intersection_b)}")
print(f"  User B - Union Size                       : {len(union_b)}")
print(f"  User B - Jaccard Similarity               : {jaccard_b:.2f}%")
print(f"  User A Dealbreakers                       : {dealbreakers_a}")
```
