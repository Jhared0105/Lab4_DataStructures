```python
# ADDITIONAL EXERCISE 3: Stateful Sliding Window Filter
# Student: Dungca | ID: TUPM-25-1053
# ============================================================
 
from collections import deque, Counter
 
# --- Required Student Variables ---
SEED_NUM        = 3
LOCAL_HAZARD    = "EARTHQUAKE"
CONTROL_NUM     = max(3, SEED_NUM)  # = 3
 
print("=== STUDENT VARIABLES ===")
print(f"LOCAL_HAZARD : {LOCAL_HAZARD}")
print(f"CONTROL_NUM  : {CONTROL_NUM}")
print()
 
# --- Initialize deque with maxlen = 5 ---
rolling_buffer = deque(maxlen=5)
 
# --- Base Burst of Alerts ---
burst = [
    (CONTROL_NUM,     "WEATHER"),
    (CONTROL_NUM + 1, "TRAFFIC"),
    (CONTROL_NUM + 2, "WEATHER"),
    (CONTROL_NUM + 3, "WEATHER"),
    (CONTROL_NUM + 4, LOCAL_HAZARD),
    (CONTROL_NUM + 5, "WEATHER"),
    (CONTROL_NUM + 6, "TRAFFIC"),
]
 
# --- Data Expansion: Append exactly (CONTROL_NUM + 2) = 5 custom alerts ---
# Timestamps continue sequentially from CONTROL_NUM + 7 onward
custom_alerts = [
    (CONTROL_NUM + 7,  LOCAL_HAZARD),
    (CONTROL_NUM + 8,  "TRAFFIC"),
    (CONTROL_NUM + 9,  LOCAL_HAZARD),
    (CONTROL_NUM + 10, "WEATHER"),
    (CONTROL_NUM + 11, "TRAFFIC"),
]
burst.extend(custom_alerts)
 
print(f"=== EXPANDED BURST LIST (Total: {len(burst)} alerts) ===")
for alert in burst:
    print(f"  {alert}")
print()
 
# --- Algorithm: Sliding Window Filter ---
print("=== SLIDING WINDOW PROCESSING ===")
rejected_count = 0
accepted_count = 0
 
for alert in burst:
    timestamp, category = alert
 
    # Extract current categories from deque using list comprehension
    current_categories = [item[1] for item in rolling_buffer]
 
    # Count category frequencies in current window
    freq_counter = Counter(current_categories)
 
    # Spam Rule: reject if category appears 2 or more times already
    if freq_counter[category] >= 2:
        print(f"  [REJECTED] Timestamp {timestamp} | Category: {category} "
              f"(already {freq_counter[category]}x in buffer)")
        rejected_count += 1
    else:
        rolling_buffer.append(alert)
        print(f"  [ACCEPTED] Timestamp {timestamp} | Category: {category} "
              f"| Buffer size: {len(rolling_buffer)}")
        accepted_count += 1
 
print()
 
# --- Final State of Deque ---
print("=== FINAL DEQUE STATE ===")
for item in rolling_buffer:
    print(f"  {item}")
print()
 
# --- Sum of Timestamps in Final Deque ---
timestamp_sum = sum(item[0] for item in rolling_buffer)
 
print("=== ASSESSMENT TABLE OUTPUT ===")
print(f"  CONTROL_NUM Used                        : {CONTROL_NUM}")
print(f"  LOCAL_HAZARD Used                       : {LOCAL_HAZARD}")
print(f"  Custom alerts added (CONTROL_NUM + 2)   : {CONTROL_NUM + 2}")
print(f"  Total alerts in expanded burst          : {len(burst)}")
print(f"  Total Accepted alerts                   : {accepted_count}")
print(f"  Total Rejected alerts                   : {rejected_count}")
print(f"  Final Deque Contents                    : {list(rolling_buffer)}")
print(f"  Sum of Timestamps in Final Deque        : {timestamp_sum}")
```
